```python
import json
import os

train_path = "../../ReadingBank/test/"

os.listdir("../../ReadingBank/test/")
```

```python
example_text_file = 'dataset-test-s2s-text-m3.json'
example_layout_file = 'dataset-test-s2s-layout-m3.json'

example_text_path = os.path.join(train_path, example_text_file)
example_layout_path = os.path.join(train_path, example_layout_file)
```

```python
with open(example_text_path, mode="r") as f:
    text_data_string = f.read()

first_data = text_data_string.split("\n")[0]

json_loaded = json.loads(first_data)

with open("../../ReadingBank/test_klein/dataset-test-s2s-text-m1.json", mode="w") as f:
    json.dump(json_loaded, f)


from pprint import pprint

pprint(json_loaded)
```

```python
with open(example_layout_path, mode="r") as f:
    text_data_string = f.read()

first_data = text_data_string.split("\n")[0]
json_loaded = json.loads(first_data)

with open("../../ReadingBank/test_klein/dataset-test-s2s-layout-m1.json", mode="w") as f:
    json.dump(json_loaded, f)


from pprint import pprint

pprint(json_loaded)
```
